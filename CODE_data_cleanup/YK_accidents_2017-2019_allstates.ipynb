```python
import pandas as pd
import time
```

```python
# import traffic data from original CSV
path="data/US_Accidents_June20.csv"
traffic_df = pd.read_csv(path)
```

```python
traffic_df # debug print
```

```python
traffic_df.columns # debug print
```

```python
traffic_df['Year']=traffic_df['Start_Time'].str.split("-", n = 1, expand = True)[0]
```

```python
traffic_df['Year'].isnull().values.any()
```

```python
filtered_df=traffic_df.loc[(traffic_df['Year']=='2019')|(traffic_df['Year']=='2018')|(traffic_df['Year']=='2017')].copy()
```

```python
#write to csv
filtered_df.to_csv('../data/YK_all2017-2019_accidents.csv',index=False)
```

```python
state_groupby=filtered_df.groupby(['State'])
```
